# 8

If we want to do a query that gives the SOPInstanceUID of all files with a measurement with name ‘PPVItd’ made on them, we could do a query like:

```python
# TODO: fix instanceidk datatype in table to avoid casting?
query = """
select a.studyid, a.name, c.SOPInstanceUID 
from measurement_abstract_rpt as a,
a_measgraphref as b,
a_instance as c
where a.name like 'PPVItd%%'
and a.studyid = b.studyidk
and a.measabstractnumber = b.measabstractnumber
and cast(b.instanceidk as text)=cast(c.instanceidk as text)
limit 5;
"""

pd.read_sql(query, conn)
```
