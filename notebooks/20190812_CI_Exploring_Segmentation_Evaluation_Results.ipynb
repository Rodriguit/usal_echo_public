## Numbers of records in the evaluation table

```python
print('{} records in the evaluation table'.format(evaluations.shape[0]))

#Count the number of instances where gt_view_name != pred_view_name
count = 0
evaluations['views dont match'] = 0
for i in evaluations.index:
    if evaluations.at[i, 'gt_view_name'] != evaluations.at[i, 'pred_view_name']:
        count +=1
        evaluations.at[i, 'views dont match'] = 1
print('{} record where the view does not match'.format(count))
print("checking column for record where the view does not match: {}".format(evaluations['views dont match'].sum()))
```

## Average IOU Scores for data subsets

```python
#for views that do match, what is the average IOU
evaluations_matching_views = evaluations[evaluations['views dont match'] == 0]
print('Average IOU for tests where the views do match: {}'.format(evaluations_matching_views['score_value'].mean()))
evaluations_non_matching_views = evaluations[evaluations['views dont match'] == 1]
print('Average IOU for tests where the views dont match: {}'.format(evaluations_non_matching_views['score_value'].mean()))
print('')
#for subsets
evaluations_matching_views_a4c = evaluations_matching_views[evaluations_matching_views['gt_view_name'] == 'a4c']
print('Average IOU for a4c tests where the views do match: {}'.format(evaluations_matching_views_a4c['score_value'].mean()))
evaluations_matching_views_a2c = evaluations_matching_views[evaluations_matching_views['gt_view_name'] == 'a2c']
print('Average IOU for a2c tests where the views do match: {}'.format(evaluations_matching_views_a2c['score_value'].mean()))
```
